```python
def simple_interest(p,n,r):
    print("simple interest calculation start")
    i = (p*n*r)/100
    a = i+p
    print(f"simple interest:{i:.2f}INR")
    print(f"amount :{a:.2f}INR")
    print(f"function is completed")
```

```python
simple_interest(1000,5,4.2)
```

```python
%%time
simple_interest(p= 200, n =3 , r= 6.3)
```

```python

def hypro(a,b):
    print("start assign value:")
    time.sleep(2)
    c = (a**2 + b**2)**(1/2)
    print(f"finally fuction  {a} and {b}:{c:.4f}INR")
    print(f"task completed")
```

```python
%%time
hypro(4,7)
```

```python
from threading import Thread
```

```python
## create thread and start its.
th1 =Thread(target=simple_interest,args=(400,2,3.2))
th2 = Thread(target = hypro,args=(2,3))

print(f"start thread")
th1.start()
th2.start()

# join thread
th1.join()
th2.join()
```

## create thread for square ..

```python
def square_mul(b:list):
    # create thread.
    threads =[]
    for i in b:
      
      th = Thread(target=square, args=(i,))
      th.start()
    threads.append(th)
    for th in threads:
      th.join()
```

```python
b1 = [10,20,30]
```

```python
b1
```

```python
%%time
square_mul(b1)
```

```python
from urllib.request import urlretrieve
```

```python
def download_file(url:str):
    filename = url.split("/")[-1]
    print(f"{filename} download start...")
    urlretrieve(url,filename)
    print(f"{filename} download complte")
```

```python
urls
```

```python
%%time
for i in urls:
    download_file(i)
    
```

```python
def multithread_download(urls:list[str]):
    threads=[]
    for i in urls:
     th = Thread(target= download_file , args =(i,))
     th.start()
     threads.append(th)
     for th in threads:
        th.join()
```

```python
%%time
multithread_download(urls)
```
